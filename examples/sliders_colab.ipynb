# SLIDERS — Colab Quickstart

Run SLIDERS on a small set of example research papers in under two minutes. Bring your own `OPENAI_API_KEY`.

**What this notebook does:**
1. Clones the public repo and installs dependencies.
2. Prompts you for an OpenAI API key (stored only in this runtime).
3. Runs SLIDERS on the bundled `sample_docs/` to answer a question across two example clinical trials.
4. Lets you ask your own question next.

Runtime cost is a few cents at current OpenAI prices.

## 2. Configure your OpenAI key

```python
import getpass, os
os.environ['OPENAI_API_KEY'] = getpass.getpass('OpenAI API key: ')
```

## 3. Run SLIDERS on the bundled example papers

`sample_docs/` contains two synthetic clinical-trial write-ups (a blood-pressure trial and a heart-failure trial). We'll ask SLIDERS to summarize and compare them.

```python
from sliders.run import run_sliders

answer = run_sliders(
    docs='sample_docs/',
    question='Summarize each trial in one sentence, then compare their primary endpoints and effect sizes.',
    openai_api_key=os.environ['OPENAI_API_KEY'],
)

print(answer)
```

## 4. Try your own question

You can either:
- change the `question` above and re-run the cell, or
- upload your own `.pdf` / `.md` files to `my_papers/` and point `docs` at that directory — SLIDERS auto-converts PDFs via Docling.

For debug artifacts (intermediate tables, extraction rationales, provenance), pass `debug=True` and `return_full_result=True`.

```python
result = run_sliders(
    docs='sample_docs/',
    question='Which trial had the larger sample size and what was its safety profile?',
    openai_api_key=os.environ['OPENAI_API_KEY'],
    debug=True,
    return_full_result=True,
)
print(result['answer'])
print('\nFull results saved to:', result['results_json_path'])
```
